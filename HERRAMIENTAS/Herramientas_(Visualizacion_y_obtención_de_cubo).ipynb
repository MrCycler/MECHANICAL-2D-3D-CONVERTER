```
# Uso de la libreria binvox to 3 array 
with open('Eje.binvox', 'rb') as f:
  model = binvox_rw.read_as_coord_array(f)
```

```
model.data.shape
```

```
x = model.data[0]
y = model.data[1]
z = model.data[2]
```

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```
# Previsualizacion de las coordenadas 
ax = plt.axes(projection='3d')
ax.set_xlim([0,256])
ax.set_ylim([0,256])
ax.set_zlim([0,256])
ax.plot_trisurf(x, y, z,
                cmap='viridis', edgecolor='none');
```
